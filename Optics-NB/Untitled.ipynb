```python
import matplotlib.pyplot as plt
from IPython.display import Latex, display
from IPython.html.widgets import interactive
from IPython.html.widgets import FloatSliderWidget, IntSliderWidget
%pylab inline --no-import-all
```

```python
from scipy import sqrt, linspace, cos, sin, pi, arctan, arcsin
from numpy import rad2deg, square,array, sum
```

```python
def FP(Tmax=1,F=1):
    phi=linspace(0,4*pi,500)
    Ttot=Tmax/(1+F*sin(phi)**2)
    fig, axes = plt.subplots(1,1)

    axes.plot(phi, Ttot, label =r'$T_{tot}')
    axes.set_ylim(0,1)
```

```python
x=interactive(FP,Tmax=(0.01,1.,),F=(.1,1000.,10))
display(x)
```

```python
def FP_resolution(deltaFrac=.01,F=10,Tmax=1):
    distance=linspace(0,.5,500)
    phi=4*pi*distance
    deltaPhi=4*pi*deltaFrac*distance
    Ttot=Tmax/(1+F*sin(phi)**2)
    Ttot1=Tmax/(1+F*sin(phi+deltaPhi)**2)
    fig, axes = plt.subplots(1,1)

    axes.plot(phi, Ttot, label =r'$T_{tot}')
    axes.plot(phi, Ttot1,label=r'$T_{tot}_1')
    axes.set_ylim(0,1.1)
    axes.set_xlim(2,4)
```

```python
x=interactive(FP_resolution,deltaFrac=(0,1,.01),Tmax=(0.01,1.,),F=(.1,1000.,10))
display(x)
```
