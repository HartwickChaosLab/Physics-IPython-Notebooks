## Setup of wavefuntions, operators, and density matricies


We will be working in the computational basis

$|0\rangle=\begin{pmatrix}1\\0\end{pmatrix}, |1\rangle=\begin{pmatrix}0\\1\end{pmatrix}$

Use np.outer to create density matricies in the computational basis.

```python
Zero=np.array([1,0])
One=np.array([0,1])
Zero_Zero=np.outer(Zero,Zero)
One_One=np.outer(One,One)
Zero_One=np.outer(Zero,One)
One_Zero=np.outer(One,Zero)
prob=np.linspace(0,1,500)
```

Define the maximally entanglled Bell States and their density matricies
http://en.wikipedia.org/wiki/Bell_state 

```python
PsiM=1/np.sqrt(2)*np.array([0,1,-1,0])
PsiP=1/np.sqrt(2)*np.array([0,1,1,0])
PhiM=1/np.sqrt(2)*np.array([1,0,0,-1])
PhiP=1/np.sqrt(2)*np.array([1,0,0,1])

PsiM_PsiM = np.outer(PsiM,PsiM)
PsiP_PsiP = np.outer(PsiP,PsiP)
PhiM_PhiM = np.outer(PhiM,PhiM)
PhiP_PhiP = np.outer(PhiP,PhiP)
```

Define some useful 2x2 matricies, namely the identity and pauli matricies.


```python
Eye=np.array([[1,0],[0,1]])
S_X=np.array([[0,1],[1,0]])
S_Y=np.array([[0,-1j],[1j,0]])
S_Z=np.array([[1,0],[0,-1]])
```

Arrays for taking partial traces.
$$
\begin{align}
Tr_{A0}&=\mathbb{1} \otimes |0\rangle\\
Tr_{A1}&=\mathbb{1} \otimes |1\rangle\\
Tr_{B0}&=|0\rangle \otimes \mathbb{1}\\
Tr_{B1}&=|1\rangle \otimes \mathbb{1}\\
\end{align}
$$

```python
TrA0=np.array([[1,0],[0,0],[0,1],[0,0]])
TrA1=np.array([[0,0],[1,0],[0,0],[0,1]])
TrB0=np.array([[1,0],[0,1],[0,0],[0,0]])
TrB1=np.array([[0,0],[0,0],[1,0],[0,1]])
```

# Quantum Transmission

To describe the transmission channel. 
$T\rho_{AB}=\sum_i E_i \rho E_i^{\dagger},$

where $\rho_{AB}$ is the total density matrix for Alice and Bob's qubit and the $E_i$s are the so-called Kraus Operators. http://en.wikipedia.org/wiki/Quantum_operation#Kraus_operators



# Quantum Transmission with Phase Noise

In this situation we have that the sent qubit $|0\rangle$ is sent faithfully to Bob, but the qubit $|1\rangle$ preserves its phase with probability $p$ and undergoes a $\pi$ phase shift with probability $(1-p)$. Since nothing happens to Alice's qubit we form the following Kraus Operators.

$$\begin{align}
E1B&=\mathbb{1} \otimes \mathbb{1}\sqrt{p}\\
E2B&=\mathbb{1} \otimes \sqrt{1-p} \sigma_z
\end{align}
$$
Where $\otimes$ is the outer product defined as below and uses np.kron in numpy.

$
\mathbf{a} \otimes \mathbf{b} = \mathbf{a} \mathbf{b}^\mathrm{T} =
\begin{bmatrix}a_1 \\ a_2 \\ a_3 \\ a_4\end{bmatrix}
\begin{bmatrix}b_1 & b_2 & b_3\end{bmatrix} =
\begin{bmatrix}a_1b_1 & a_1b_2 & a_1b_3 \\ a_2b_1 & a_2b_2 & u_2b_3 \\ a_3b_1 & a_3b_2 & a_3b_3 \\ a_4b_1 & a_4b_2 & a_4b_3\end{bmatrix}.$

## Quantum Transmission with Amplitude Damping

In this situation we have that the sent qubit $|0\rangle$ is sent faithfully to Bob, but the qubit $|1\rangle$ preserves its state with probability $1-p$ and decays to $|0\rangle$ with probability $(p)$. Since nothing happens to Alice's qubit we form the following Kraus Operators.

$$\begin{align}
E1B&=|0\rangle\langle 0|+\sqrt{1-p}|1\rangle\langle 1|\\
E2B&=\sqrt{p}|0\rangle\langle 1|
\end{align}$$

```python
def Transmission(rho_in,p,noisetype):
    if noisetype==0: #0 corresponds to phase noise
        E1B = np.kron(Eye,np.sqrt(1-p)*Eye)
        E2B = np.kron(Eye,np.sqrt(p)*S_Z)
    else: #1 corresponds to Amplitude Damping
        E1B=np.kron(Eye,Zero_Zero+np.sqrt(1-p)*One_One)
        E2B=np.kron(Eye,np.sqrt(p)*Zero_One)
        
    return np.dot(np.dot(E1B,rho_in),E1B.T)+np.dot(np.dot(E2B,rho_in),E2B.T)
```

## Partial Trace

The function Partial_Trace 'traces out' Alice or Bob. The partial trace corresponds to measuring one qubit without knowing the results. 

In the notation of above definitions

$\rho_A=Tr_{A0}^\dagger\rho_{AB}Tr_{A0}+Tr_{A1}^\dagger\rho_{AB}Tr_{A1}$

```python
def Partial_Trace(rho_in,who):
    if who==0:
        return np.dot(np.dot(TrA0.T,rho_in),TrA0)+ np.dot(np.dot(TrA1.T,rho_in),TrA1)
    else:
        return np.dot(np.dot(TrB0.T,rho_in),TrB0)+ np.dot(np.dot(TrB1.T,rho_in),TrB1)
```

As an example, take one of the maximally mixed Bell States and 'trace out' Alice. In other words, "Given his state, can Bob figure out what Bell State he and Alice share?" Since we traced out Alice, we don't have the results from her, so we expect Bob won't know what Bell State he had.

```python
Partial_Trace(PsiM_PsiM,1)
```

We see that if Bob were to now make a measurement on his qubit, he would have a 50-50 chance of getting a $|0\rangle$ or $|1\rangle$, so he can't figure out which Bell State he shared.

## Information Entropy

Claude Shannon defined Information as the amount of surprise contained in a message. The less probable a message, the more information it contains. A weather report stating that Los Angeles will have a mild day with sunshine is not that surprising, so it doesn't convey much information. Mathematically we write the informations as $I(M)=1/p(M)$ Where M is a random message and $p(M)$ is its probability.

The entropy of a message is defined as the average amount of inforamtion conveyed in a message and is given by $H(X)=-\sum_{j=1}^N p_j \log_2(p_j)$

In Quantum Mechanics we define the entropy a little differently. If we encode the messages as qubits $|x_j\rangle$ and each has a probabability $p_j$ to be sent, we define the message's density operator as $\rho=\sum_j p_j |x_j\rangle \langle x_k|$. From this we can define the entropy of a quantum message.

$S(\rho)=-Tr{\rho \log_2(\rho)}=\sum_j p_j \log_2(p_j)$

Where we exploited the orthogonality of the messages.

## Joint Entropy

The Joint Entropy is the total entropy contained in Bob's and Alice's qubits. It is defined as $S(\rho_{AB})=-Tr[\rho_{AB} \log_2[\rho_{AB}]]$, where $\rho_{AB}$ is the combined density operators of Alice's and Bob's qubits. $\rho_{AB}=\rho_A \otimes \rho_B$

```python
def Joint_Entropy(rho_in, p,noisetype):
    rho_AB=Transmission(rho_in,p,noisetype)
    eigenvalues= [val for val in linalg.eigvals(rho_AB) if val !=0]
    #return eigenvalues
    return -np.sum(np.real(eigenvalues)*np.log2(np.real(eigenvalues)))
```

## Single Entropy

To find the entropy of one of the qubits, we trace out the other qubit from the Joint Entropy

```python
def Single_Entropy(rho_in,p,who,noisetype):
    eigenvalues=[val for val in linalg.eigvals(Partial_Trace(Transmission(rho_in,p,noisetype),who)) if val !=0]

    return -np.sum(np.real(eigenvalues)*np.log2(np.real(eigenvalues)))
    
```

```python
SEA1=np.zeros(len(prob)) #Single Entropy for Alice under dephasing
SEA2=np.zeros(len(prob)) #Single Entropy for Alice under amplitude loss
SEB1=np.zeros(len(prob)) #Single Entropy for Bob under dephasing
SEB2=np.zeros(len(prob)) #Single Entropy for Bob under amplitude loss

for i in range(0,len(prob)):
    SEA1[i]=Single_Entropy(PsiM_PsiM, prob[i], 0,0)
    SEA2[i]=Single_Entropy(PsiM_PsiM, prob[i], 0,1)
    SEB1[i]=Single_Entropy(PsiM_PsiM, prob[i], 1,0)
    SEB2[i]=Single_Entropy(PsiM_PsiM, prob[i], 1,1)

rcdef = plt.rcParams.copy()
newparams = {'axes.labelsize': 14, 'axes.linewidth': 1, 'savefig.dpi': 300, 
             'lines.linewidth': 1.5, 'figure.figsize': (16, 6),
             'figure.subplot.wspace': 0.4,
             'ytick.labelsize': 12, 'xtick.labelsize': 12,
             'ytick.major.pad': 5, 'xtick.major.pad': 5,
             'legend.fontsize': 12, 'legend.frameon': False, 
             'legend.handlelength': 1.5}

# Update the global rcParams dictionary with the new parameter choices
# Before doing this, we reset rcParams to its default again, just in case
plt.rcParams.update(rcdef)
plt.rcParams.update(newparams)

# Make the new figure with new formatting
fig, axes = plt.subplots(1, 2)

axes[0].plot(prob, SEA1, label ='dephasing')
axes[0].plot(prob,SEA2, label='amplitude loss')
axes[1].plot(prob, SEB1, label='dephasing')
axes[1].plot(prob,SEB2, label='amplitude loss')

axes[0].legend()
axes[1].legend()
axes[0].set_title('''Alice's Entropy''')
axes[0].set_xlabel('probability')
axes[0].set_ylabel('bits')
axes[1].set_title('''Bob's Entropy''')
axes[1].set_xlabel('probability')
axes[1].set_ylabel('bits')
plt.rcParams.update(rcdef)
```

## Conditional Entropy

Conditional entropy measures how much entropy a random message X has remaining after we have received the message Y. If Y and X are uncorrelated, then we the conditional entropy is the same as the entropy of X, and Bob learns no new information about the message. On the other hand, if there is perfect correlation between the sent and received messages, then when Bob measures Y, he gains perfect information S(Y) and the uncertainty of X goes to zero and so does the conditional entropy. The idea of quantum conditional entropy seems to be a bit problematic, since QM doesn't have a formal definition of conditional probability; however, for bipartate systems the following definition works well:

$S(A|B)_\rho=S(\rho_{AB})-S(B)_\rho $ 

In quantum mechanics this quantitity can be negative. The interpretation of negative conditional entropy is that this is a measure of the excess number of bits above the classical limit that can be transmitted  via a quantum protocol. 

```python
def Conditional_Entropy(rho_in,p,noisetype):
    return Joint_Entropy(rho_in,p,noisetype)-Single_Entropy(rho_in,p,1,noisetype)
    
    
```

## Mutual Information

The mutual information tells us the information that two random variables X and Y share, where X is the sent message and Y is the received message. If X and Y are uncorrelated, then the knowledge of Y tells us nothing about the message X. On the other hand if there two are completely correlated, then knowledge of Y tells us all about the sent message. Mathematically we write it as:

$M(\rho_{AB})=S(A)_\rho+S(B)_\rho-S(\rho_{AB})$

```python
def Mutual_Information(rho_in,p,noisetype):
    return Single_Entropy(rho_in,p,0,noisetype)+Single_Entropy(rho_in,p,1,noisetype)-Joint_Entropy(rho_in,p,noisetype)
```

We see that QM allows us to send 2 bits of information with only 1 qubit, but as the probability of amplitude loss goes up, we pass through the classical limit and we see that we wind up doing worse.
