# Exerise 2.2

Use circular motion law with Newton's gravity law. Or, use Kepler's third law directly. $$F_g=\frac{mv^2}{(R+h)}\\$$
$$\frac{GmM}{(R+h)^2}=\frac{mv^2}{(R+h)}$$
For a circular orbit, $v=\frac{2\pi(R+h)}{T}$ so
$$\frac{GMT^2}{4\pi^2}=(R+h)^3$$
Then it is trivial to get the rest of the answer

```python
from math import pow,pi
#First define some constants.
G=6.67e-11 #m^3kg^-1s^-2
M=5.97e24 #kg
R=6371e3 #m
#Easier for me to think in minutes
Tmin=float(input("Enter the desired orbital period in minutes: "))
T=Tmin*60 #convert to seconds
h=pow((G*M*T**2)/(4*pi**2),1./3)-R
print("The desired orbital height over the Earth is ",h/1000,"km")
```

# Exercise 2.4
Remember that $\gamma=1/\sqrt{1-v^2/c^2}$ We are told rest frame is on Earth so $T_{earth}=x/v$. The distance the spaceship goes with respect to an observer on Earth is given by $x/\gamma$ so $T_{ship}=x/v\gamma$. The problem asks for very specific units, so watch what you put in for $c$ for $\gamma$

```python
from math import sqrt

d=float(input("Enter distance in light years"))
v=float(input("Enter speed in units of speed of light"))
gamma=1/sqrt(1-v**2) #c=1 in our units.
print("The elapsed time on Earth: ", d/v, "years")
print("Mission control says the time elapsed for the astronauts is: ",d/(v*gamma), "years")
```

# Exercise 2.5
Fairly easy just watch units. An easy check is that to within rounding T+R better equal 1

```python
from math import sqrt

hbar=6.582119514e-16 #units of ev s
m=511e3 #ev/c^2
c=299792458 #m/s
KE=float(input("Enter Kinetic Energy of particle in eV:"))
V=float(input("Potential Barrier Energy in eV (please make less than KE)"))
k1=sqrt(2*m*c**2*KE)/hbar
k2=sqrt(2*m*c**2*(KE-V))/hbar
T=4*k1*k2/(k1+k2)**2
R=(k1-k2)**2/(k1+k2)**2
print("the transmission is ",round(T,2)," and the reflection is ",round(R,2))
```

# Exercise 2.6


```python
from math import sqrt, pi

M=1.9891e30 #kg
G=6.6738e-11 #m^3/Kg/s^2
l1=float(input("Please enter perihelion distance in meters"))
v1=float(input("Please enter perihelion speed in m/s"))
# Form constants for solving quadratic
B=-2*G*M/(v1*l1)
C=-(v1**2-2*G*M/l1)
#taking smallest root
v2=(-B-sqrt(B**2-4*C))/2
l2=l1*v1/v2
a=(l1+l2)/2
b=sqrt(l1*l2)
T=2*pi*a*b/(l1*v1)
e=(l2-l1)/(l2+l1)
print("The aphelion distance is", l2,"m")
print("The aphelion velocity is", v2,"m/s")
print("The semi-major axis is ", round(a,2),"m")
print("The semi-minor axis is ", round(b,2),"m")
print("The orbital period is ", round(T,2), "s")
print("The orbital period in years",T/(24*3600*365))
print("The orbital eccentricity is", round(e,2))
```

# Exercise 2.9: Madelung Constant

Set different values of L and pick i, j, k that are between -L and L+1. Check to make sure i=j=k=0 doesn't happen, then check for i+j+k = even or odd


```python
from math import sqrt

L=500
M=0.0
for i in range (-L,L+1):
    for j in range (-L,L+1):
        for k in range (-L,L+1):
            if i==0 and j==0 and k ==0:
                continue
            if (i+j+k)%2==0:   #even
                M+= 1/sqrt(i*i+j*j+k*k)
            else:
                M-= 1/sqrt(i*i+j*j+k*k)
print(M)
```

# Exercise 2.11

write two functions one for factorial and one for binomial

```python
def factorial(n): #assumes n is an integer. Very inelegant failure if not
    f=1.0
    
    if n==0:
        return 1 #check for n=0
    else:
        for k in range(1,n+1):
            f*=k
    return f
```

```python
def binomial(n,k):
    if k==0:
        return 1
    else:
        binomial=factorial(n)/(factorial(k)*factorial(n-k))
    return binomial
```

```python
binomial(4,2)
```

Pascal's Triangle

```python
nmax=20

for n in range(1,nmax+1):
    pascal_line=[]
    for k in range(0,n+1):
        pascal_line.append(binomial(n,k))
    print(pascal_line)
```

```python
def probability(n,k):
    return binomial(n,k)/2**n
```

```python
probability(100,60)
```

```python
# probability that coin comes up heads more than 60 times in a row. As a check kmin=n should give within rounding 1
n=100
kmin=50
cumlProb=0
for k in range(kmin, n+1):
    cumlProb+=probability(n,k)
print(cumlProb)
```

# Exercise 2.12
Follow the suggestions. Remember to cast $\sqrt(n)$ to int.

```python
from math import sqrt
primes=[2]
for n in range(3,500,2): #skip by two, b/c we know even numbers are not prime excepting 2.
    stop=int(sqrt(n))
    k=0
    while primes[k]<=stop:
        if n%primes[k]==0:
            break
        k+=1
    else:
        primes.append(n)
print(primes)
```

# Exercise 2.13

```python
#factorial with recursion This will override earlier factorial
def factorial(n):
    if n==1:
        return 1
    else:
        return n*factorial(n-1)
```

```python
factorial(5)
```

```python
#recursive catalan
def catalan(n):
    if n==0:
        return 1
    else:
        return (4*n-2)/(n+1)*catalan(n-1)
```

```python
for i in range(0,1000000):
    print(catalan(i))
```

```python
#gcd
def gcd(m,n):
    if n==0:
        return m
    else:
        return gcd(n,m%n)
```

```python
gcd(108,192)
```
