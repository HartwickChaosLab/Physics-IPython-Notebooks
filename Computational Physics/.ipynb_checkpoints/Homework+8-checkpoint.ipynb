```python
from numpy import arange,array, sin, cos
from pylab import plot, show, xlabel, ylabel
```

```python
#8.1 A Low Pass Filter

def vin(t):
    if (2*t//1)%2==0:
        v=1
    else:
        v=-1
    return v
```

```python
def f(vo,t):
    return (vin(t)-vo)/RC
```

```python

a = 0.0
b = 10.0
N = 1000
h = (b-a)/N


x=0.0
vo=0.0
rc=[1.0,0.1,0.01]

for RC in rc:
    tpoints = arange(a,b,h)
    xpoints = []
    for t in tpoints:
        xpoints.append(x)
        k1 = h*f(x,t)
        k2 = h*f(x+0.5*k1,t+0.5*h)
        k3 = h*f(x+0.5*k2,t+0.5*h)
        k4 = h*f(x+k3,t+h)
        x += (k1+2*k2+2*k3+k4)/6

    plot(tpoints,xpoints)
    xlabel("t")
    ylabel("x(t)")
show()
```

```python
#8.3 The Lorentz Equations
```

```python
def f(r,t):
    x=r[0]
    y=r[1]
    z=r[2]
    fx=S*(y-x)
    fy=R*x-y-x*z
    fz=x*y-z*B
    return array ([fx,fy,fz],float)

a=0.0
b=50.0
N=10000
h=(b-a)/N

tpoints=arange(a,b,h)
xpoints=[]
ypoints=[]
zpoints=[]

S=10
R=28
B=8/3

r=array([0.0,1.0,0.0],float)
for t in tpoints:
    xpoints.append(r[0])
    ypoints.append(r[1])
    zpoints.append(r[2])
    
    k1=h*f(r,t)
    k2=h*f(r+0.5*k1,t+0.5*h)
    k3=h*f(r+0.5*k2,t+0.5*h)
    k4=h*f(r+k3,t+h)
    r+=(k1+2*k2+2*k3+k4)/6

plot(tpoints,ypoints,"gold")
xlabel("t")
ylabel("y(t)")
show()
    
```

```python
#The Driven Pendulum

def f2(r,t):
    theta=r[0]
    omega=r[1]
    ftheta=omega
    fomega=-(g/l)*sin(theta)+C*cos(theta)*sin(ohm*t)
    return array([ftheta,fomega],float)
```

```python
a=0.0
b=100.0
N=1000
h=(b-a)/N

g=9.81
l=.1
C=2
ohm=10

tpoints=arange(a,b,h)
xpoints=[]
ypoints=[]

theta=0
omega=0

r=array([0.0,0.0],float)
for t in tpoints:
    xpoints.append(r[0])
    ypoints.append(r[1])
    
    k1=h*f2(r,t)
    k2=h*f2(r+0.5*k1,t+0.5*h)
    k3=h*f2(r+0.5*k2,t+0.5*h)
    k4=h*f2(r+k3,t+h)
    r+=(k1+2*k2+2*k3+k4)/6

plot(xpoints,tpoints,"gold")
xlabel("x")
ylabel("t")
show()
```
