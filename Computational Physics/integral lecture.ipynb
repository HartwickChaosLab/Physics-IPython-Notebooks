```python
from numpy import array,sin, cos
```

```python
def function(x):
    return (x**4-2*x+1)
```

```python
def integrate_trap(slices, a,b):
    h=(b-a)/slices
    integral=0.5*(function(a)+function(b))
    for k in range(1,slices):
        integral+=function(a+k*h)
    return(integral*h)
```

```python
integrate_trap(10,0,2)
```

```python
function(3)
```

```python
num=[10,100,1000,10000]

for slices in num:
    
    estimate=integrate_trap(slices,0,3)
    percent_error=(3.-estimate)/3.*100
    print("Slices: ",slices, " integral= ",estimate, " percent error=",round(percent_error,8))
```

```python
int_1=integrate_trap(10,0,2)
int_2=integrate_trap(20,0,2)
print("integral 1, ",int_1," integral 2, ",int_2," error, ",(int_2-int_1)/3.)
```

```python
def integrate_simp(slices, a,b):
    h=(b-a)/slices
    
    int_e=0
    int_o=0
    for k in range(1,slices,2):
        int_o+=function(a+k*h)
    for k in range(2,slices,2):
        int_e+=function(a+k*h)
    integral=function(a)+function(b)+2*int_e+4*int_o
    return(integral*h/3)
```

```python
integrate_simp(10,0,2)
```

```python
int_1=integrate_simp(10,0,2)
int_2=integrate_simp(20,0,2)
print("integral 1, ",int_1," integral 2, ",int_2," error, ",(int_2-int_1)/12.)
```

Adaptive integration

```python
def adapt_trap(error,N,a,b):
    
    diff=2*error
    while (abs(diff)>error):
        int_1=integrate_trap(N,a,b)
        int_2=integrate_trap(2*N,a,b)
        diff=(int_2-int_1)/3.
        N=N*2
        #print("integral 1, ",int_1," integral 2, ",int_2," error, ",diff)
    print(int_2)
```
